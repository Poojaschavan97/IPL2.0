### Plan of action

- creating a dataframe containing
    - team1
    - team2
    - season
    - city
    - venue
    - winner
    - target runs of both teams
    - inning
    

```python
# Step 1: Calculate total runs for each match and inning
inning_runs = deliveries.groupby(['match_id', 'inning'])['total_runs'].sum().reset_index()

# Step 2: Separate into first and second innings
inning1_runs = inning_runs[inning_runs['inning'] == 1][['match_id', 'total_runs']].rename(columns={'total_runs': 'team1_runs'})
inning2_runs = inning_runs[inning_runs['inning'] == 2][['match_id', 'total_runs']].rename(columns={'total_runs': 'team2_runs'})

# Step 3: Merge with matches DataFrame
matches = matches.merge(inning1_runs, on='match_id', how='left')
matches = matches.merge(inning2_runs, on='match_id', how='left')

# Display the updated DataFrame
matches.head()
```

```python
dashboard_data=matches[['match_id','team1','team2','city','venue','season','team1_runs','team2_runs','winner']]
```

```python
dashboard_data=dashboard_data.dropna()
```
